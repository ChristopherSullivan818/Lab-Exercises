```python
import requests
import tkinter as tk
from tkinter import *


#GUI #1
class MyQuotes:
    def __init__(self):
        self.root = Tk()
        self.root.title('Author quotes')
        self.root.geometry('900x500')
       
        self.dropdown_label = tk.Label(self.root, text='Choose a person:', font=('Ariel', 15))
        self.dropdown_label.pack(pady=10)
       
        self.dropdown = ttk.Combobox(self.root, values=[], state='readonly')
        self.dropdown.pack()
       
        self.label = tk.Label(self.root, text='')
        self.label.pack(pady=10)
       
        self.get_button = tk.Button(self.root, text='Quote',font=('Ariel', 11), command=self.get_quote)
        self.get_button.pack(pady=10)
       
        self.reset_button = tk.Button(self.root, text='Reset', command=self.reset)
        self.reset_button.pack(pady=10)
       
        self.update_dropdown()
   
    def update_dropdown(self):
        self.dropdown.config(values=[])
        response = requests.get('https://api.quotable.io/authors')
        if response.status_code == 200:
            authors = [author['name'] for author in response.json()['results']]
            self.dropdown.config(values=authors)
   
    def get_quote(self):
        author = self.dropdown.get()
        if author:
            response = requests.get(f'https://api.quotable.io/quotes?author={author}&limit=1')
            if response.status_code == 200:
                quote = response.json()['results'][0]['content']
                self.label.config(text=f'"{quote}" - {author}')
   
    def reset(self): #reset button
        self.dropdown.set('')
        self.label.config(text='')
        self.update_dropdown()

if __name__ == '__main__':
    window = MyQuotes()
    window.root.mainloop()

    
```
